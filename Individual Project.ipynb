```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv("mlb-umpire-scorecard.csv")
data
```

```python
"""Finding the average umpire accuracy"""
data = data[data["accuracy"].str.contains("ND") == False]
```

```python
data['accuracy'] = pd.to_numeric(data['accuracy'])
```

```python
avgAcc = data['accuracy'].mean()
avgAcc
```

```python
data.hist(column='accuracy')
```

```python
"""Finding the average number of runs by which the home team is favored by the umpire"""
data['favor_home'] = pd.to_numeric(data['favor_home'])
```

```python
avgHomeFav = data['favor_home'].mean()
avgHomeFav
```

```python
data.hist(column='favor_home')
```

```python
"""Finding the total run impact by an umpire on a game"""
data['total_run_impact'] = pd.to_numeric(data['total_run_impact'])
```

```python
avgImpact = data['total_run_impact'].mean()
avgImpact
```

```python
data.hist(column='total_run_impact')
```

```python
"""Finding the difference between accuracy and expected accuracy"""
data['expected_accuracy'] = pd.to_numeric(data['expected_accuracy'])
data['accuracy_above_expected'] = pd.to_numeric(data['accuracy_above_expected'])
```

```python
avgAccDiff = data['accuracy_above_expected'].mean()
avgAccDiff
```

```python
data.hist(column='accuracy_above_expected')
```

```python
"""Finding plot of consistency"""
data['consistency'] = pd.to_numeric(data['consistency'])
```

```python
data.hist(column='consistency')
```

```python
yr_2018 = data.loc[data['date'].str.contains("2018", case=False)]
yr_2018
```

```python
yr_2017 = data.loc[data['date'].str.contains("2017", case=False)]
yr_2017
```

```python
yr_2016 = data.loc[data['date'].str.contains("2016", case=False)]
yr_2016
```

```python
yr_2015 = data.loc[data['date'].str.contains("2015", case=False)]
yr_2015
```

```python
acc_2015 = yr_2015['accuracy'].mean()
acc_2016 = yr_2016['accuracy'].mean()
acc_2017 = yr_2017['accuracy'].mean()
acc_2018 = yr_2018['accuracy'].mean()
acc_2019 = yr_2019['accuracy'].mean()
acc_2020 = yr_2020['accuracy'].mean()
acc_2021 = yr_2021['accuracy'].mean()
acc_2022 = yr_2022['accuracy'].mean()
```

```python
year_data = [[2015, acc_2015], [2016, acc_2016], [2017, acc_2017], [2018, acc_2018], [2019, acc_2019], [2020, acc_2020], 
             [2021, acc_2021], [2022, acc_2022]]
year_df = pd.DataFrame(year_data, columns=['Year', 'Average Umpire Accuracy'])
year_df
```

```python
year_df.plot.bar(x='Year', y='Average Umpire Accuracy', title='Average Umpire Accuracy by Year', legend=None)
```

```python
"""Plotting average accuracy above expected for each year"""
aae_2015 = yr_2015['accuracy_above_expected'].mean()
aae_2016 = yr_2016['accuracy_above_expected'].mean()
aae_2017 = yr_2017['accuracy_above_expected'].mean()
aae_2018 = yr_2018['accuracy_above_expected'].mean()
aae_2019 = yr_2019['accuracy_above_expected'].mean()
aae_2020 = yr_2020['accuracy_above_expected'].mean()
aae_2021 = yr_2021['accuracy_above_expected'].mean()
aae_2022 = yr_2022['accuracy_above_expected'].mean()
```

```python
year_data2 = [[2015, aae_2015], [2016, aae_2016], [2017, aae_2017], [2018, aae_2018], [2019, aae_2019], [2020, aae_2020], 
             [2021, aae_2021], [2022, aae_2022]]
year_df2 = pd.DataFrame(year_data2, columns=['Year', 'Average Umpire Accuracy Above Expected'])
year_df2
```

```python
year_df2.plot.bar(x='Year', y='Average Umpire Accuracy Above Expected', 
                  title='Average Umpire Accuracy Above Expected by Year', legend=None)
```

```python
"""Plotting average consistency for each year"""
con_2015 = yr_2015['consistency'].mean()
con_2016 = yr_2016['consistency'].mean()
con_2017 = yr_2017['consistency'].mean()
con_2018 = yr_2018['consistency'].mean()
con_2019 = yr_2019['consistency'].mean()
con_2020 = yr_2020['consistency'].mean()
con_2021 = yr_2021['consistency'].mean()
con_2022 = yr_2022['consistency'].mean()
```
